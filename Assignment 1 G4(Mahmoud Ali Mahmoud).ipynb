```python
import networkx as nx
import numpy as np
from networkx.algorithms import tree
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

```python
Gedges=np.array([[0,0,4,4,0,0],
                 [0,0,0,3,3,0],
                 [4,0,0,2,0,0],
                 [4,3,2,0,4,2],
                 [0,3,0,4,0,3],
                 [0,0,0,2,3,0]
                ])
```

```python
Gedges
```

```python
G = nx.from_numpy_array(Gedges)
G.edges(data=True)
int2label = {0: "A", 1: "B", 2: "C", 3: "D", 4: "E", 5: "F"}
G = nx.relabel_nodes(G, int2label)
weights = nx.get_edge_attributes(G, "weight")
pos=nx.spiral_layout(G)
nx.draw_networkx_edge_labels(G, pos, edge_labels=weights)
nx.draw_networkx(G,pos)
plt.show
```

```python
G = nx.Graph() 
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F'])
G.add_edge('A', 'C',weight=4)
G.add_edge('A', 'D',weight=4)
G.add_edge('D', 'B',weight=3)
G.add_edge('D', 'E',weight=4)
G.add_edge('D', 'F',weight=2)
G.add_edge('C', 'D',weight=2)
G.add_edge('F', 'E',weight=3)
G.add_edge('B', 'E',weight=3)
```

```python
pos=nx.spring_layout(G) 
nx.draw_networkx(G,pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
plt.show()
```

### Deapth first search

```python
depthSearch=nx.dfs_tree(G,"A")
nx.draw(depthSearch,with_labels=True)
```

### Breadth First Search

```python
BreadthList=list(nx.bfs_tree(G,"A"))
Breadth=nx.bfs_tree(G,"A")
nx.draw(Breadth,with_labels=True)
```

### dijkstra_path

```python
print(nx.dijkstra_path(G,source="A",target="E"))
```

### kruskal

```python
mstkruskal = tree.minimum_spanning_edges(G, algorithm="kruskal", data=False)
list(mstkruskal)
```

### prim

```python
mstprim = tree.minimum_spanning_edges(G, algorithm="prim", data=False)
list(mstprim)
```
